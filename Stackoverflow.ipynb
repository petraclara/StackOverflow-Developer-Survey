```python
def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []

    for idx, value in col_series[col_series.notnull()].items():
        for option in value.split(';'):
            if option not in result_df.columns:
                options.append(option)
                result_df[option] = False
            result_df.at[idx, option] = True

    return result_df[options]
```

```python
employment_type_df= split_multicolumn(survey_raw_df.Employment)
```

```python
employment_type_totals= employment_type_df.sum().sort_values(ascending = False)
```

## At what companies do developers get paid the most?

```python

# Selecting relevant columns
relevant_columns = ['OrgSize','EdLevel', 'ConvertedCompYearly', 'Country', 'Currency']

# Creating a new DataFrame with only the relevant columns
salary_data = survey_raw_df[relevant_columns]

# Displaying the first few rows of the DataFrame
print(salary_data.head())
```

```python
print(salary_data['ConvertedCompYearly'].describe())
```

```python
# Boxplot
sns.boxplot(x='OrgSize', y='ConvertedCompYearly', data=salary_data)
plt.xticks(rotation=45)
plt.xlabel('Organization Size')
plt.ylabel('Converted Compensation (Yearly)')
plt.title('Yearly Converted Compensation Across Organization Sizes')
plt.show()
```

```python
# Grouping by organization size and calculating average salary
avg_salary_by_org = salary_data.groupby('OrgSize')['ConvertedCompYearly'].mean().sort_values(ascending=False)

print(avg_salary_by_org)
```

```python

avg_salary_by_org = salary_data.groupby('OrgSize')['ConvertedCompYearly'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_salary_by_org.index, y=avg_salary_by_org.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Organization Size')
plt.ylabel('Average Converted Compensation (Yearly)')
plt.title('Average Yearly Converted Compensation by Organization Size')
plt.show()
```

## How does coding experience affect the level of pay?

```python
# Group by YearsCode and calculate the average salary
average_salary_by_experience = survey_raw_df.groupby('YearsCode')['ConvertedCompYearly'].mean().reset_index()

```

```python
average_salary_by_experience_range
```

```python
# Define custom bins for grouping
bins = [0, 2, 5, 10, 15, 20, 25, 30, 35, 40, 50, 60, float('inf')]  # Customize the bin edges as needed

# Create labels for the bins
labels = ['0-2', '3-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-50', '51-60', '60+']

# Add a new column with the experience range
survey_raw_df['ExperienceRange'] = pd.cut(survey_raw_df['YearsCode'], bins=bins, labels=labels, right=False)

# Group by the new column and calculate the average salary
average_salary_by_experience_range = survey_raw_df.groupby('ExperienceRange')['ConvertedCompYearly'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='ExperienceRange', y='ConvertedCompYearly', data=average_salary_by_experience_range)
plt.title('Average Salary by Years of Coding Experience Range')
plt.xlabel('Years of Coding Experience Range')
plt.ylabel('Average Salary (Yearly)')
plt.show()
```
